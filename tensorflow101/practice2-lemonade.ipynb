# 레모네이드 판매 예측
- github csv url: https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv 

```python
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```python
# 데이터를 준비합니다.
파일경로 = '../data/csv/lemonade.csv'
레모네이드 = pd.read_csv(파일경로)
레모네이드.head()
```

```python
# 종속변수, 독립변수
독립 = 레모네이드[['온도']]
종속 = 레모네이드[['판매량']]
print(독립.shape, 종속.shape)
```

```python
# 모델을 만듭니다.
X = tf.keras.layers.Input(shape=[1])    # 독립변수의 갯수 1
Y = tf.keras.layers.Dense(1)(X)         # 종속변수의 갯수 1 = 종속변수는 항상 1개이다. 마지막 층!
model = tf.keras.models.Model(X, Y)     # 모델에 input층과 마지막 층 추가
model.compile(loss='mse')               # 컴파일: 손실함수 mse
```

```python
# 모델을 학습시킵니다. 
model.fit(독립, 종속, epochs=10000, verbose=0)  # verbose=0 화면 출력 없이
model.fit(독립, 종속, epochs=10)                # 화면 출력해서 손실값 확인
```

```python
# 모델을 이용합니다. 
print(model.predict(독립))
print(종속)
```

```python
# 새로운 값 입력해서 예측해보기
print(model.predict([[15]]))
```
